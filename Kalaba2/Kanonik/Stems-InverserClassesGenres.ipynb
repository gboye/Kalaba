# Inverser le genre et les classes des noms

```python
from os.path import expanduser
import yaml,re
```

```python
home = expanduser("~")
repertoire=home+"/Github/Kalaba/Kalaba2/Kanonik/"
repertoire=home+"/ownCloud/Cours/Bordeaux/L1-LinguistiqueGenerale/00-ProjetKalaba/23-K5/"

with open(repertoire+"Stems.yaml", 'r') as stream:
    prevStems=yaml.load(stream,Loader=yaml.Loader)
    
newStems=prevStems.copy()
```

```python
if "N1" in prevStems["NOM"].keys():
    cfNouns={}
    genderCf={}
    for cf in prevStems["NOM"].keys():
        if isinstance(prevStems["NOM"][cf],dict):
            for g in prevStems["NOM"][cf].keys():
                cfNouns[cf]=[]
                for k,v in prevStems["NOM"][cf][g].iteritems():
                    cfNouns[cf].append({k:v})
                    if not g in genderCf:
                        genderCf[g]={}
                    genderCf[g][cf]=cfNouns[cf]
genderCf
```

```python
newStems["NOM"]=genderCf
```

```python
newStems
```

```python
with open(repertoire+'newStems.yaml', 'w') as output:
    yaml.dump(newStems, output, default_flow_style=False,allow_unicode=True)

with open(repertoire+"newStems.yaml", 'r') as input:
    yamlLines=input.readlines()    
    
yamlText="".join(yamlLines)
yamlText=re.sub(r"\n\s*-\s*",", ",yamlText)
```

```python
yamlText=re.sub(r"\n(\s*)([^:\n]*)\s*:,\s*([^:\n]*)\s*:","\n\g<1>\g<2>:\n\g<1> \g<3>:",yamlText)
# print yamlText
```

```python
def repairLigne(text):
    return re.sub(r"\n( *)([^:\n]*) *:([^:\n]*) +(\S+) *:","\n\g<1>\g<2>:\g<3>\n\g<1>\g<4>: ",text)

prevText=""
newText=yamlText
while prevText!=newText:
    prevText=newText
    newText=repairLigne(prevText)
yamlText=newText
# print yamlText
```

```python
yamlText=re.sub(r":\s*,\s*",": ",yamlText)
# print yamlText
```

```python
yamlText=re.sub(r": *([^\n]+)",": [\g<1>]",yamlText)
# print yamlText
```

```python
with open(repertoire+'newStems.yaml', 'w') as output:
    output.write(yamlText)
```
