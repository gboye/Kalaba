```python
import yaml
with open(repKanoniks+"K08-NomAcc/Stems.yaml") as inFile:
    stems=yaml.safe_load(inFile)
```

```python
def imprimeNiveau(dNiveau):
    for k,v in dNiveau.iteritems():
        if isinstance(v,dict):
            print "\t\\item",k
            imprimeNiveau(v)
        elif isinstance(v,list):
            print "\t\\begin{itemize}"
            print "\t\t\\item",k
            print "\t\t\\begin{itemize}"
            print "\t\t\t\\item", ", ".join(v)
            print "\t\t\\end{itemize}"
            print "\t\\end{itemize}"
        else:
            print "\t\\begin{itemize}"
            print "\t\t\\item",k
            print "\t\t\\begin{itemize}"
            print "\t\t\t\\item", v
            print "\t\t\\end{itemize}"
            print "\t\\end{itemize}"
```

```python
sortie=[]
```

```python
sortie.append("\\begin{itemize}")
```

```python
sortie.append("\t\\item DET")
sortie.append("\t\\begin{description}")
for k,v in stems["DET"].iteritems():
    sortie.append("\t\t\\item [%s] %s"%(k,", ".join(v)))
sortie.append("\t\\end{description}")
```

```python
sortie.append("\t\\item PRO")
sortie.append("\t\\begin{description}")
for k,v in stems["PRO"].iteritems():
    sortie.append("\t\t\\item [%s] %s"%(k,", ".join(v)))
sortie.append("\t\\end{description}")
```

```python
sortie.append("\t\\item NOM")
sortie.append("\t\\begin{itemize}")
for kC,vC in stems["NOM"].iteritems():
    sortie.append("\t\t\\item "+kC)
    sortie.append("\t\t\\begin{itemize}")
    for kG,vG in vC.iteritems():
        sortie.append("\t\t\t\\item "+kG)
        sortie.append("\t\t\t\\begin{description}")
        for k,v in vG.iteritems():
            sortie.append("\t\t\t\t\\item [%s] %s"%(k,", ".join(v)))
        sortie.append("\t\t\t\\end{description}")
    sortie.append("\t\t\\end{itemize}")
sortie.append("\t\\end{itemize}")
```

```python
sortie.append("\t\\item ADJ")
sortie.append("\t\\begin{description}")
for k,v in stems["ADJ"].iteritems():
    sortie.append("\t\t\\item [%s] %s"%(k,", ".join(v)))
sortie.append("\t\\end{description}")
```

```python
sortie.append("\t\\item VER")
sortie.append("\t\\begin{description}")
for k,v in stems["VER"].iteritems():
    sortie.append("\t\t\\item [%s] %s"%(k,", ".join(v)))
sortie.append("\t\\end{description}")
```

```python
sortie.append("\t\\item PREP")
sortie.append("\t\\begin{description}")
for k,v in stems["PREP"].iteritems():
    sortie.append("\t\t\\item [%s] %s"%(k,", ".join(v)))
sortie.append("\t\\end{description}")
```
