```python
import re,yaml,codecs
from os.path import expanduser
import pandas as pd
from cellbell import ding
```

```python
home = expanduser("~")
annee="2020"
numGroupe=50
repertoire=home+"/ownCloud/Cours/Bordeaux/L1-UE4-Morphologie/00-Kanonik/"+annee
repertoire=home+"/Github/Kalaba/Kalaba2/Kanonik/"
```

```python
nomPhrases=repertoire+"NewKalaba.csv"
with codecs.open(nomPhrases,encoding="cp1252") as inFile:
    lignes=[l.strip() for l in inFile.readlines()[1:]]
# lignes
```

```python
formes={}
for ligne in lignes:
    mots=[m for m in re.split("[,; '."+'"]',ligne) if m!=""]
    for mot in mots:
        formes[mot.lower()]=[]
formes
```

```python
nomLexique=home+"/ownCloud/Cours/Bordeaux/L2-XML/XML-Ressources/bdlexique.txt"
dfLexique=pd.read_csv(nomLexique,encoding="utf8",sep=";")
```

```python
dfLexique.columns="ortho phono liaison cat ms tam lemme L23 F1 F2 F3 F4".split()
catLexique={"NOM":"N G".split(),"VER":"V".split(),"ADJ":"F G J K L M".split()}
tam="pi ii ai".split()
pers="3S 3P".split()
```

```python
catGroups={}
groupCats={"VER":["K","V"],"NOM":["F","J","N"], "ADJ":["A","G","M"],"PREP":["P"],"DET":["d","p"]}
for cat in groupCats:
    for groupCat in groupCats[cat]:
        catGroups[groupCat]=cat
```

```python
vocabulaire={}
for forme in formes:
    for lemme in dfLexique[dfLexique["ortho"]==forme]["lemme"].tolist():
        fCats=dfLexique[dfLexique["lemme"]==lemme]["cat"].tolist()
        for fCat in fCats:
            gCat=catGroups[fCat]
            if gCat not in vocabulaire:
                vocabulaire[gCat]={}
            if lemme not in vocabulaire[gCat]:
                vocabulaire[gCat][lemme]=set()
            vocabulaire[gCat][lemme].add(forme)
for cat in vocabulaire:
    for lemme in vocabulaire[cat]:
        vocabulaire[cat][lemme]=list(vocabulaire[cat][lemme])
```

```python
with open(repertoire+"Vocabulaire.yaml", 'w') as stream:
    yaml.safe_dump(vocabulaire,stream=stream,default_flow_style=False,allow_unicode=True)
```
